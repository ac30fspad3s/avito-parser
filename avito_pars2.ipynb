```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
```

```python
#считывааем html страницу и ищем на ней ссылки на объявления
flat_links=[]
for i in range(20):
    time.sleep(5)
    r = requests.get(f'https://www.avito.ru/tver/kvartiry/prodam-ASgBAgICAUSSA8YQ?cd=1&p=i')
    #print(r)
    soup = BeautifulSoup(r.text, 'html.parser')
    flats = soup.find_all('a', {'class': 'link-link-39EVK link-design-default-2sPEv title-root-395AQ iva-item-title-1Rmmj title-listRedesign-3RaU2 title-root_maxHeight-3obWc'})
    for j in range(len(flats)):
        flat_links.append("https://www.avito.ru" + flats[j]['href'])
```

```python
for i in range(len(flat_links)):
    print(flat_links[i])
    print()
```

```python
#Заполняем информацию об объявлениях (Адрес, район, тип дома,тип предложения, этаж, кол-во этажей, площадь, жилая площадь, пложадь кухни, балкон, цена)
id = 0
data = []
for i in range(len(flat_links)):
    id = id + 1
    pge = flat_links[i]
    r = requests.get(pge)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    if soup.find('span', {'class': 'item-address__string'}) is not None:
        address  = soup.find('span', {'class': 'item-address__string'}).text
    else:
        #address = " "
        continue
    
    if soup.find('span', {'class': 'item-address-georeferences-item__content'}) is not None:
        district = soup.find('span', {'class': 'item-address-georeferences-item__content'}).text
    else: 
        #address = " "
        continue
        
        
    #balcony = " "
    ##type_h = soup.find('span', {'class': 'item-params-label'}) 
    ##count_rooms = soup.find('span', {'class': 'item-params-label'}) 
    ##floor = soup.find('span', {'class': 'item-address-georeferences-item__content'}) 
    ##all_square = soup.find('span', {'class': 'item-address-georeferences-item__content'}) 
    ##kitchen_square = soup.find('span', {'class': 'item-address-georeferences-item__content'}) 
    ##square = soup.find('span', {'class': 'item-address-georeferences-item__content'}) 
    ##balcony = soup.find('span', {'class': 'item-address-georeferences-item__content'}) 
    header = soup.find_all('span', itemprop="name")
    market_type = " "
    for i in range(len(header)):
        if header[i].text == "Вторичка":
            market_type = "Вторичка"
        elif header[i].text == "Новостройки":
            market_type = "Новострой"
    
    price = soup.find('span', {'class': 'js-item-price'})['content']
    
    params = soup.find_all('li', {'class': 'item-params-list-item'})
    for j in range(len(params)):
    #print(params[i].text)
        buff = params[j].text.split(':')
    #print(buff[0] == " Тип дома")
        if buff[0] == " Тип дома":
        #print("+")
            type_h = buff[1]
        elif buff[0] == " Этаж":
            floor_str = buff[1].split("из")
            floor = floor_str[0]
            count_floor = floor_str[1]
        elif buff[0] == " Количество комнат":
            count_rooms = buff[1]
        elif buff[0] == " Общая площадь":
            square_str = buff[1].split("\xa0м²")
            all_square = square_str[0]
        elif buff[0] == " Жилая площадь":
            square_str = buff[1].split("\xa0м²")
            square = square_str[0]
        elif buff[0] == " Площадь кухни":
            square_str = buff[1].split("\xa0м²")
            kitchen_square = square_str[0]
        elif buff[0] == " Балкон или лоджия":
            balcony = buff[1]
    
    data.append([id, address.strip() , district.strip() , type_h.strip() , floor.strip() , count_floor.strip() , count_rooms.strip() , all_square.strip() , square.strip() , kitchen_square.strip() , balcony.strip() , market_type.strip() , price.strip() ])
    print(id)
    time.sleep(5)
```

```python
#запись в файл
FILENAME = "Tver_flats_14_05_2021.csv"
with open(FILENAME, "w", encoding="utf-8", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
